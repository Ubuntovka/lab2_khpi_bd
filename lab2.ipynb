# Исследование csv файла

```python
list_reading_csv_time = []
list_adding_csv_time = []
```

```python
import os


file_csv_stats = os.stat("healthy_lifestyle_city_2021.csv")

print(f'CSV file size in MegaBytes is {file_csv_stats.st_size / (1024 * 1024)}')
```

## Ячейка для выводов

Перед добавлением размер был:

После добавления данных:

Файл увеличился потому что данные добавились.

```python
import csv
import time


start_time = time.time()

# adding row in csv file
with open('healthy_lifestyle_city_2021.csv', 'a') as f:
    writer = csv.writer(f)
    writer.writerow(['Kharkiv', '10', '1938', '£1.08', '29.40%', '81.70', '24.26', '1670', '7.23', '218', '788', '£31.04'])
    
```

```python
if adding_csv_time not in list_adding_csv_time:
    list_adding_csv_time.append(adding_csv_time)

print("--- %s seconds ---" % adding_csv_time)
```

```python
import time
import csv

# reading csv file

start_time = time.time()

with open('healthy_lifestyle_city_2021.csv', newline='') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)

reading_csv_time = time.time() - start_time
```

```python
if reading_csv_time not in list_reading_csv_time:
    list_reading_csv_time.append(reading_csv_time)

print("--- %s seconds ---" % reading_csv_time)
```



# Синий график про чтение, желтый про добавление



```python
%matplotlib notebook
import matplotlib.pyplot as plt

x = [int(i) for i in range(len(list_reading_csv_time))]
plt.xlabel("Номер попытки")         # ось абсцисс
plt.ylabel("Время выполнения")    # ось ординат
plt.plot(x, list_reading_csv_time, x, list_adding_csv_time)
plt.show()
```

```python
average_reading_csv = sum(list_reading_csv_time) / (len(list_reading_csv_time))
print(f'Среднее время чтения файла: {average_reading_csv}')

average_adding_csv = sum(list_adding_csv_time) / (len(list_adding_csv_time))
print(f'Среднее время добавления в файл: {average_adding_csv}\n')

print(f"Худшее время чтения файла: {max(list_reading_csv_time)}")
print(f"Худшее время добавления в файл: {max(list_adding_csv_time)}")
```

# Исследование бинарного файла

```python
list_reading_bin_time = []
list_adding_bin_time = []
```

```python
import os


file_csv_stats = os.stat("data.bin")

print(f'BIN file size in MegaBytes is {file_csv_stats.st_size / (1024 * 1024)}')
```

## Ячейка для выводов

Перед добавлением размер был:

После добавления данных:

Файл увеличился потому что данные добавились.

```python
import time


start_time = time.time()

# adding data in bin file
with open('data.bin', "ab+") as f:
    t = ("azxsdcvfgb" + "\n").encode()
    f.write(t)

adding_bin_time = time.time() - start_time

if adding_bin_time not in list_adding_bin_time:
    list_adding_bin_time.append(adding_bin_time)

print("--- %s seconds ---" % adding_bin_time)
```

```python
import time


start_time = time.time()

# reading bin file
with open('data.bin', "rb+") as f:
    for line in f:
        print(line.decode())

reading_bin_time = time.time() - start_time
```

```python
if reading_bin_time not in list_reading_bin_time:
    list_reading_bin_time.append(reading_bin_time)

print("--- %s seconds ---" % reading_bin_time)
```

# Синий график про чтение, жёлтый про добавление

```python
%matplotlib notebook
import matplotlib.pyplot as plt

x = [int(i) for i in range(len(list_reading_bin_time))]
plt.xlabel("Номер попытки")         # ось абсцисс
plt.ylabel("Время выполнения")    # ось ординат
plt.plot(x, list_reading_bin_time, x, list_adding_bin_time)
plt.show()
```

```python
average_reading_bin = sum(list_reading_bin_time) / (len(list_reading_bin_time))
print(f'Среднее время чтения файла: {average_reading_bin}')

average_adding_bin = sum(list_adding_bin_time) / (len(list_adding_bin_time))
print(f'Среднее время добавления в файл: {average_adding_bin}\n')

print(f"Худшее время чтения файла: {max(list_reading_bin_time)}")
print(f"Худшее время добавления в файл: {max(list_adding_bin_time)}")
```
